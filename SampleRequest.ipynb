```python
# import from '.env' file
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
import requests

url = "https://data.nayaone.com/credit_card_data"

payload = {}
headers = {
  'Accept-Profile': 'api',
  'sandpit-key': os.getenv('API_KEY')
}

response = requests.request("GET", url, headers=headers, data=payload)

print(response.text)
```
