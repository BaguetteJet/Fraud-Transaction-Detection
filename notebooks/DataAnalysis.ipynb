```python
import pandas as pd
import matplotlib.pyplot as plt
```

Load raw data

```python
df = pd.read_csv('../data/raw/credit_data.csv')
df.info()
df.describe()
```

Check for duplicate data

```python
df.duplicated().any()
```

```python
df.boxplot()
df.boxplot(figsize=(12, 200))
plt.xticks(rotation=45)
plt.show()
```

```python
df.hist(figsize=(12, 10), bins=30)
plt.show()
```

Check class imbalance

```python
# count values of classifier column
counts = df['fraud'].value_counts()
counts
```

```python
# pie chart
counts.plot(kind='pie', labels=['Genuine', 'Fraud'], autopct='%1.1f%%')
plt.title('Class Distribution')
plt.show()
```

If classes are imbalanced, avoid Accuracy metric, use F1 Score, Percision and Recall instead
